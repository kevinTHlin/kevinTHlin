```python
from scipy import stats

df3_total = df3.iloc[315]
factor_name = ["形考成绩", "提交作业次数", "阶段性测验成绩", "作业时长", "学习表现成绩", "记分作业成绩", "课程浏览次数", "课程总浏览时间", "作业正确率", "课程视频时长"] 
four = np.array(df3_total[0:10])
five = np.array(df3_total[10:20])
seven = np.array(df3_total[20:])
data = {"Ranking_before_seeing_the_algorithm": four, "Ranking_after_seeing_the_algorithm": five, "Ranking_for_causal_reasoning": seven}

df5 = pd.DataFrame(data, columns = ["Ranking_before_seeing_the_algorithm", "Ranking_after_seeing_the_algorithm", "Ranking_for_causal_reasoning"], index= factor_name)
df5
```

```python
obs = np.array([df3_total[0:10], df3_total[10:20]])
stats.chi2_contingency(obs)
#Since the p-vale is less than .05, we concluded that there is a significant relationship 
#between the rank (i.e., the one variable) and before/after seeing the algorithm (i.e., the other varibel). 
#This means the effect of algorithm aversion does not exist!
```

```python
obs2 = np.array([df3_total[10:20], df3_total[20:]])
stats.chi2_contingency(obs2)
#Since the p-vale is larger than .05, we concluded that there is no significant relationship 
#between the rank (i.e., the one variable) and correlational/causal reasoning (i.e., the other varibel). 
#This means the effect of false causal inference fallacy exists! 
```
