```python
features = New_df2[New_df2.columns[~New_df2.columns.isin(["TOTALSCORE"])]]
features
```

```python
target = New_df2["TOTALSCORE"]
target
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.model_selection import KFold, cross_val_score, StratifiedKFold, train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import AdaBoostClassifier
```

```python
standardizer = StandardScaler()

RF = RandomForestClassifier()
logit = LogisticRegression()
SVM = LinearSVC()
AB = AdaBoostClassifier()

pipeline_RF = make_pipeline(standardizer, RF)
pipeline_logit = make_pipeline(standardizer, logit)
pipeline_SVM = make_pipeline(standardizer, SVM)
pipeline_AB = make_pipeline(standardizer, AB)
pipeline_set = [pipeline_RF, pipeline_logit, pipeline_SVM, pipeline_AB]

kf = StratifiedKFold(n_splits=6, shuffle=True, random_state=1)

scoring = {'acc': 'accuracy',
           'prec_macro': 'precision_macro',
           'rec_micro': 'recall_macro'}

features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.1, random_state=1)
```

```python
#Is there a better way???
for i in pipeline_set:
    print(cross_val_score(i, 
                    features_train, 
                    target_train, 
                    cv = kf, 
                    scoring = "accuracy", 
                    n_jobs=-1))
```

```python
# why it did not work???
print(for i in pipeline_set:
      cross_val_score(i, 
                      features_train, 
                      target_train, 
                      cv = kf, 
                      scoring = "accuracy", 
                      n_jobs=-1))
```

```python
#why the values are changed (different from the last cell) even when set as "random_state=1"
model_name = ["RF", "logit", "SVM", "AB"]

for i,j in zip(pipeline_set, model_name): 
    print(ny.mean(cross_val_score(i, 
                    features_train, 
                    target_train, 
                    cv = kf, 
                    scoring = "accuracy", 
                    n_jobs=-1)))
    print(j)
```

```python
from sklearn.preprocessing import LabelBinarizer   #the labls were binary already; why do we still need this process for recall, precision, and f1?

lb = LabelBinarizer()
for i,j in zip(pipeline_set, model_name): 
    print(ny.mean(cross_val_score(i, 
                    features_train, 
                    lb.fit_transform(target_train), 
                    cv = kf, 
                    scoring = "precision", 
                    n_jobs=-1)))
    print(j)
```

```python
#Is there a better way for precision, recall and f1 at the same time???
for i,j in zip(pipeline_set, model_name): 
    print(ny.mean(cross_val_score(i, 
                    features_train, 
                    lb.fit_transform(target_train), 
                    cv = kf, 
                    scoring = "recall", 
                    n_jobs=-1)))
    print(j)
```
